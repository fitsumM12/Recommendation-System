# data cleaning and preprocessing

```python
# removing  the records for which the id is not avaliable
movie_md = movie_md[movie_md['id'].str.isnumeric()]
```

```python
# merge the dataframe as a single entity
# to merge the  ids must be of the same   datatypes
movie_md['id']= movie_md['id'].astype(int)
# Merge
df = pd.merge(movie_md, movie_keywords, on ="id", how='left')
df = pd.merge(df, movie_credits, on ='id', how='left')
```

```python
# reset the index
df.reset_index(inplace=True,drop=True)
```

```python
df.head()
```

```python
# lets fetch the genres, keywords, cast to vectorize them later
# lets first start with cleaning the movies metadata
# fetching the genre list from the columns
df['genres']=df['genres'].apply(lambda x: [i['name'] for i in eval(x)])
```

```python
# replaces spaces in between genre(ex - sci  fi to  scifi) and make it a string
df['genres']=df['genres'].apply(lambda x: ' '. join([i.replace(" ", "")for i in x]))
```

```python
# filling the num values as []
df['keywords'].fillna('[]', inplace=True)
```

```python
# lets clean  the keywords dataframe to extract the keywords
# fetching the keywords list from the columns
df['keywords']=df['keywords'].apply(lambda x:[i['name']  for i in eval(x)])
```

```python
# remove the empty space and join all the keywords with spaces
df['keywords'] = df['keywords'].apply(lambda x: ' '.join([i.replace(" ","")for i in x]))
```

```python
# fill the num values as []
df['cast'].fillna('[]', inplace = True)
```

```python
# lets clean the cast dataframe to extract the name of the actors from cast colums
# fetching the cast list from the columns
df['cast'] = df['cast'].apply(lambda x: [i['name'] for i in eval(x)])
# remove the empty spaces and join all the cast with spaces
df['cast']= df['cast'].apply(lambda x: ' '.join([i.replace(" ", '') for i in x]))
```

```python
# lets merge all content of the movies as a single feature
df['tags'] = df['overview']+' '+df['genres']+' '+df['original_title']+' '+ df['keywords']+' '+df['cast']
```

# Model based collaborative filtering 

```python
# read the movies mmeta-data
movie_md = pd.read_csv("C:/Users/KIIT01/MTECH/DATASET/movie_dataset/movies_metadata.csv")
# read the keywords
movie_keywords = pd.read_csv("C:/Users/KIIT01/MTECH/DATASET/movie_dataset/keywords.csv")
# read the credits
movie_credits = pd.read_csv("C:/Users/KIIT01/MTECH/DATASET/movie_dataset/credits.csv")
ratings = pd.read_csv("C:/Users/KIIT01/MTECH/DATASET/movie_dataset/ratings_small.csv")
```

```python
ratings.head()
```

```python
# movie dataframe with votes more than 55
movie_md = movie_md[movie_md['vote_count']>55][['id','title']]
```

```python
# IDS of movies with count more than 55
movie_ids = [int(x) for x in movie_md['id'].values]
# select ratings of movies with more than 55 counts
ratings = ratings[ratings['movieId'].isin(movie_ids)]
```

```python
# Reset Index
ratings.reset_index(inplace=True,drop=True)
# prints first 5 rows
ratings.head()
```

```python
ratings.shape
```

```python
# Initialize a surprise
from surprise import Dataset,Reader
```

```python
from surprise.prediction_algorithms.matrix_factorization import SVD
from surprise import accuracy
```

```python
# Intialize the surprise reader object
reader = Reader(line_format="user item rating", sep=',', rating_scale=(0,5),skip_lines = 1)
```

```python
# load the data
data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader=reader)
```

```python
# Build trainset object(perform this only when you are using whole dataset to train)
trainset = data.build_full_trainset()
```

```python
# Initialize Model
svd = SVD()
# cross-validate
svd.fit(trainset)
```

```python
#  we have fit the data successfully, now lets check some prediction
svd.predict(uid=3,iid=2959,r_ui=5)
```

```python
svd.predict(uid=15,iid=2678, r_ui=1)
```

```python
def get_recommendations(data, movie_md, user_id, top_n, algo):
    # creating an empty list to store the recommended product ids
    recommendations =[]
    # creating  an user item interactiions matrix
    user_movie_interactions_matrix = data.pivot(index='userId',columns = 'movieId', values ="rating")
    # extracting those product ids which the user_id has not interacted yet
    non_interacted_movies = user_movie_interactions_matrix.loc[user_id][user_movie_interactions_matrix.loc[user_id].isnull()].index.tolist()
    # looping through each of the product ids which user_id has not interacted yet
    for item_id in non_interacted_movies:
        # predicting the ratings for those non interacted product ids by this user
        est = algo.predict(user_id, item_id).est
        #appending the predicted ratings
        movie_name = movie_md[movie_md['id']==str(item_id)]['title'].values[0]
        recommendations.append((movie_name,est))
    recommendations.sort(key = lambda x:x[1],reverse=True)
    return recommendations[:top_n]
```

```python
get_recommendations(data=ratings, movie_md = movie_md, user_id=654, top_n=10, algo=svd)
```

# Memory Based Recommender System
### user-Based

```python
# The memory based method use user rating historical data to compute
# the similarity between users or items.
# the idea behind these methods is to define a similarity measure between users
# or items and find the most similar to recommend unseen items. memory based recommender 
# systems are two types-
# 1. user-based
# 2. item-based
```

```python
from surprise.prediction_algorithms.knns import KNNBasic
# declare the similarity options
sim_options ={'name':'cosine',
             'user_based':True}
# knn algorithm is used  to find  the similarity items
sim_user = KNNBasic(sim_options = sim_options , verbose=False, random_state=33)
# train the algorithm on the trainset, and predict ratings for the testset
sim_user.fit(trainset)
```

```python
# predicting rating for a sample user with an interacted product
sim_user.predict(uid=2, iid=17, r_ui=5)
```

```python
# predicting rating for sample user with an interacting product
sim_user.predict(uid=671, iid=4011, r_ui=4)
```

```python
get_recommendations(ratings, movie_md, 671, 10, sim_user)
```

### Item-Based

```python
# declare the similarity options
sim_options = {'name':'cosine',
              'user_based':False}
# knn algorithm is used to find the similarity items
sim_item = KNNBasic(sim_options = sim_options, verbose=False, random_state=33)
# train the algorithm on the trainset and predict
sim_item.fit(trainset)
```

```python
# predicting rating for a sample user with an interacted product
sim_item.predict(uid=671, iid=4011, r_ui=4)
```

```python
get_recommendations(ratings,movie_md, 671, 10, sim_item)
```
